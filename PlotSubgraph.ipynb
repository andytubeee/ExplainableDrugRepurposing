```python
i = 0
```

```python
g_single_instance, g_neg_single_instance, u, v = get_single_instance_graph(g, etype, i)
```

```python
with open(f'Output/Explainability/Alzheimer/imp_node_dict_{i}_{mode}.json') as file:
    imp_node_dic = json.load(file)
```

```python
with open('Output/Explainability/Alzheimer/node_mapping.json') as file:
    node_dict = json.load(file)
```

```python
compound_id = list(node_dict[etype[0]].keys())[u]
disease_id = list(node_dict[etype[2]].keys())[v]
compound_name, disease_name = get_drug_disease_names(compound_id, disease_id)

print(f'Drug name: {compound_name}')
print(f'Disease name: {disease_name}')
```

**Map important nodes to its label**

```python
label_imp_nodes = []
label_imp_nodes.append(compound_id)
label_imp_nodes.append(disease_id)
for key in keys:
    lst = imp_node_dic[key]
    for x in lst:
        label_imp_nodes.append(list(node_dict[key].keys())[x])
```

**Get explanatory subgraph**

```python
df = pd.read_csv('Input/DRKG/drkg.tsv', sep='\t', header=None)
df = df.dropna()
```

```python
df_explain = df[(df[0].isin(label_imp_nodes)) & (df[2].isin(label_imp_nodes))]
```

```python
G = nx.from_pandas_edgelist(df_explain, 0, 2, edge_attr=True, create_using=nx.MultiDiGraph())
```

## Plot

```python
nodelist = {}
for node in G.nodes:
    split = node.split('::')
    if split[0] in nodelist:
        nodelist[split[0]] += [node]
    else: 
        nodelist[split[0]] = [node]
```

```python
labels = {}
for node in G.nodes:
    split = node.split('::')
    labels[node] = split[len(split)-1]
```

```python
edge_labels = {}
for i, row in df_explain.iterrows():
    split = row[1].split('::')
    pair = (row[0], row[2])
    edge_labels[pair] = row[1]
```

```python
sub_g, inverse_indices = dgl.khop_out_subgraph(g, {'Compound': u.item(), 'Disease': v.item()}, k=2)
```

```python
plt.figure(figsize = (24, 12))
pos = nx.spring_layout(G, k = 2.5)

#nodes
options = {'node_size': 999, 'alpha': 0.99}
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Compound'], node_color="yellow", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Disease'], node_color="orange", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Gene'], node_color="green", **options)
#nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Biological Process'], node_color="blue", **options)
#nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Molecular Function'], node_color="violet", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Pathway'], node_color="cyan", **options)

#node labels
nx.draw_networkx_labels(G, pos, labels, font_size=12, font_color='black')

#edges
nx.draw_networkx_edges(G, pos)

#edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color='black')

#legend + title
color = ['yellow', 'orange', 'green', 'blue', 'violet', 'cyan']
label = ['Compound', 'Disease', 'Gene', 'Biological Process', 'Molecular Function', 'Pathway']
handles = [plt.plot([],[], marker="o", ms=10, ls="", mec=None, color=color[i], 
           label="{:s}".format(label[i]))[0] for i in range(len(label))]
plt.legend(handles=handles)
plt.title(f'Explanatory Subgraph for {compound_name} treats {disease_name}')
plt.show()
```
