## Define Use Case: one drug treats one disease

```python
i = 2
```

```python
g_single_instance, g_neg_single_instance, u, v = get_single_instance_graph(g, etype, i)
```

```python
with open(f'Output/Explainability/Alzheimer/imp_node_dict_{i}_{mode}.json') as file:
    imp_node_dic = json.load(file)
```

```python
with open('Output/Explainability/Alzheimer/node_mapping.json') as file:
    node_dict = json.load(file)
```

```python
compound_id = list(node_dict[etype[0]].keys())[u]
disease_id = list(node_dict[etype[2]].keys())[v]
compound_name, disease_name = get_drug_disease_names(compound_id, disease_id)

print(f'Drug name: {compound_name}')
print(f'Disease name: {disease_name}')
```

**Map important nodes to its label**

```python
label_imp_nodes = []
label_imp_nodes.append(compound_id)
label_imp_nodes.append(disease_id)
for key in keys:
    lst = imp_node_dic[key]
    for x in lst:
        label_imp_nodes.append(list(node_dict[key].keys())[x])
```

**Get explanatory subgraph - Filter on defined meta path**

```python
df = pd.read_csv('Input/DRKG/drkg.tsv', sep='\t', header=None)
df = df.dropna()
```

```python
df_explain = df[(df[0].isin(label_imp_nodes)) & (df[2].isin(label_imp_nodes))]
```

```python
df_filter = df_explain[df_explain[1].isin({'DGIDB::BINDER::Gene:Compound', 'Hetionet::CbG::Compound:Gene',
                                           'Hetionet::DaG::Disease:Gene', 'Hetionet::GpPW::Gene:Pathway'})]
df_filter = df_filter.rename(columns={0:'subject', 1:'predicate', 2:'object'})
```

```python
df_relation_glossary = pd.read_csv('Input/DRKG/relation_glossary.tsv', sep='\t')
df_relation_glossary = df_relation_glossary[['Relation-name', 'Interaction-type']]
df_relation_glossary = df_relation_glossary.rename(columns={'Relation-name':'predicate'})
df_filter = df_filter.merge(df_relation_glossary, how='left', on='predicate')
```

```python
gene_binding = list(df_filter[df_filter['Interaction-type'] == 'binding']['object'])
gene_association = list(df_filter[df_filter['Interaction-type'] == 'association']['object'])
gene_participation = list(df_filter[df_filter['Interaction-type'] == 'participation']['subject'])
genes = [value for value in gene_binding if value in gene_association and value in gene_participation]
df_filter = df_filter[df_filter['subject'].isin(genes) | df_filter['object'].isin(genes)]
```

**Add importance scores**

```python
df_scores = pd.DataFrame()
for key in keys:
    temp = pd.DataFrame()
    temp['NodeMappingID'] = imp_node_dic[key]
    temp['Score'] = imp_node_dic[key+'_score']
    temp['NodeID'] = [list(node_dict[key].keys())[x] for x in imp_node_dic[key]]
    df_scores = df_scores.append(temp)
```

```python
df_filter = df_filter.merge(df_scores[['Score', 'NodeID']], left_on='subject', right_on='NodeID')
df_filter = df_filter.rename(columns={'Score':'subject_score'})
df_filter = df_filter.drop(columns={'NodeID'})
```

```python
df_filter = df_filter.merge(df_scores[['Score', 'NodeID']], left_on='object', right_on='NodeID')
df_filter = df_filter.rename(columns={'Score':'object_score'})
df_filter = df_filter.drop(columns={'NodeID'})
```

```python
df_filter['score'] = df_filter['subject_score'] + df_filter['object_score']
```

**Rank meta paths**

```python
genes = df_scores[df_scores['NodeID'].isin(genes)].sort_values('Score', ascending=False)['NodeID']
genes = genes[:3]
```

```python
genes
```

```python
df_metapaths = pd.DataFrame()
for gene in np.unique(genes):
    temp_gene = df_filter[(df_filter['subject'] == gene) | (df_filter['object'] == gene)]
    metapath1 = temp_gene[(temp_gene['Interaction-type'] == 'binding')].sort_values('score', ascending=False).iloc[0]
    metapath2 = temp_gene[(temp_gene['Interaction-type'] == 'association')].sort_values('score', ascending=False).iloc[0]
    metapath3 = temp_gene[(temp_gene['Interaction-type'] == 'participation')].sort_values('score', ascending=False).iloc[0]
    df_metapaths = df_metapaths.append([metapath1, metapath2, metapath3])
```

```python
df_metapaths
```

**Add labels**

```python
# Pathway
df_hetionet = pd.read_csv('Input/DRKG/HETIONET_vocabulary.tsv', sep='\t')
df_hetionet = df_hetionet[df_hetionet['kind'] == 'Pathway']
dict_hetionet = dict(zip(df_hetionet['id'], df_hetionet['name']))
```

```python
# Compound
df_db = pd.read_csv('Input/DRKG/DB_vocabulary.csv')
df_db['DrugBank ID'] = 'Compound::' + df_db['DrugBank ID']
dict_db = dict(zip(df_db['DrugBank ID'], df_db['Common name']))
```

```python
# Disease
df_mesh = pd.read_csv('Input/DRKG/MESH_vocabulary.tsv', sep='\t')
df_mesh['d'] = df_mesh['d'].str.upper()
df_mesh['d'] = 'Disease::' + df_mesh['d']
dict_mesh = dict(zip(df_mesh['d'], df_mesh['name']))
```

```python
# Gene
df_ncbi_gene = pd.read_csv('Input/DRKG/NCBIGENE_vocabulary.tsv', sep='\t')
df_ncbi_gene['ncbi'] = df_ncbi_gene['ncbi'].str.replace('http://bio2rdf.org/ncbigene:','Gene::')
dict_ncbi_gene = dict(zip(df_ncbi_gene['ncbi'], df_ncbi_gene['label']))
```

```python
dict_labels = {**dict_hetionet, **dict_db, **dict_mesh, **dict_ncbi_gene}
```

```python
subject_names = []
object_names = []
for i,row in df_metapaths.iterrows():
    if row['subject'] in dict_labels: subject_names.append(dict_labels[row['subject']])
    else: subject_names.append(row['subject'])
    if row['object'] in dict_labels: object_names.append(dict_labels[row['object']])
    else: object_names.append(row['object'])
df_metapaths['subject_names'] = subject_names
df_metapaths['object_names'] = object_names
```

```python
for i,row in df_metapaths[['subject_names', 'Interaction-type', 'object_names']].iterrows():
    print(row['subject_names'] + ' ---' + row['Interaction-type'] + '---> ' + row['object_names'])
```

## Plot

```python
G = nx.from_pandas_edgelist(df_metapaths, 'subject', 'object', edge_attr=True, create_using=nx.MultiDiGraph())
```

```python
nodelist = {}
for node in G.nodes:
    split = node.split('::')
    if split[0] in nodelist:
        nodelist[split[0]] += [node]
    else: 
        nodelist[split[0]] = [node]
```

```python
labels = {}
for node in G.nodes:
    if node in dict_labels: labels[node] = dict_labels[node]
    else: labels[node] = node
```

```python
edge_labels = {}
for i, row in df_metapaths.iterrows():
    pair = (row['subject'], row['object'])
    edge_labels[pair] = row['Interaction-type']
```

```python
plt.figure(figsize = (24, 12))
pos = nx.fruchterman_reingold_layout(G, k = 2.5)

#nodes
options = {'node_size': 999, 'alpha': 0.99}
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Compound'], node_color="yellow", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Disease'], node_color="orange", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Gene'], node_color="lightgreen", **options)
nx.draw_networkx_nodes(G, pos, nodelist=nodelist['Pathway'], node_color="lightblue", **options)

#node labels
nx.draw_networkx_labels(G, pos, labels, font_size=12, font_color='black')

#edges
nx.draw_networkx_edges(G, pos)

#edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color='black')

#legend + title
color = ['yellow', 'orange', 'lightgreen', 'lightblue']
label = ['Compound', 'Disease', 'Gene', 'Pathway']
handles = [plt.plot([],[], marker="o", ms=10, ls="", mec=None, color=color[i], 
           label="{:s}".format(label[i]))[0] for i in range(len(label))]
plt.legend(handles=handles)
plt.title(f'Explanatory Subgraph for {compound_name} treats {disease_name}')
plt.show()
```
